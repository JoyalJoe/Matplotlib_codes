```python
days=[1,2,3,4,5,6,7]
max=[60,65,63,67,62,69,61]
min=[50,55,53,57,52,59,51]
avg=[55,52,58,69,57,64,56]
```

```python
# Giving labels for the plotting data

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max, label="Max")
plt.plot(days,min, label="Min")
plt.plot(days,avg, label="Avg")
plt.legend(loc="upper left")
```

```python
import matplotlib.pyplot as plt
import numpy as np

company=("EY","Cognizent","UST","Amazon","Netflix","Google")
revenue = {
    'Data': (65,95,80,45,75,55),
    'profit' : (45,55,65,75,55,105),
    
}

x = np.arange(len(company)) 
width = 0.25  
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in revenue.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1


ax.set_ylabel('Company Revenue')
ax.set_title("US Tech Stocks:")
ax.set_xticks(x + width, company,)
ax.legend(loc='best')
ax.set_ylim(0, 150)

plt.show()
```

```python
# Pie Chart

expenses=[12000,6000,1000,2000,3000,5000]
labell=["MF","Home","Travel","Shopping","Others","Saving"]
plt.pie(expenses,labels=labell)
plt.show()# To not to show writings on the chart
```

```python
# To show the percentage

expenses=[12000,6000,1000,2000,3000,5000]
labell=["MF","Home","Travel","Shopping","Others","Saving"]
plt.pie(expenses,labels=labell,autopct="%0.1f%%",shadow=True)
plt.show()# To not to show writings on the chart


# To cut the whole graph in pieces explode is used

expenses=[12000,6000,1000,2000,3000,5000]
labell=["MF","Home","Travel","Shopping","Others","Saving"]
plt.pie(expenses,labels=labell,autopct="%0.1f%%",shadow=True,
        explode=[0.2,0,0.3,0,0,0],startangle=45)
plt.show()
plt.savefig("C:/Users/ASUS/OneDrive/Pictures/piechart.png")
```
